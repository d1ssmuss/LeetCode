```python
import pandas as pd

# Данные
data = {
    'city': ['Jacksonville', 'Jacksonville', 'Jacksonville', 'Jacksonville', 'Jacksonville',
             'ElPaso', 'ElPaso', 'ElPaso', 'ElPaso', 'ElPaso'],
    'month': ['January', 'February', 'March', 'April', 'May',
              'January', 'February', 'March', 'April', 'May'],
    'temperature': [13, 23, 38, 5, 34, 20, 6, 26, 2, 43]
}

# Создание DataFrame
df = pd.DataFrame(data)

# Вывод DataFrame
df
```

```python
# Создаем pivot таблицу: строки – даты, столбцы – продукты, значения – продажи
pivot_table = df.pivot(index='month', columns='city', values='temperature')
print("Базовая pivot таблица:")
pivot_table
```
